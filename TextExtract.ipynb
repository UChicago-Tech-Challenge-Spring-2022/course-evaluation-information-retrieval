```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
HTMLFile = open(file, "r")
contents = HTMLFile.read()
bsText = BeautifulSoup(contents, 'html')
```

```python
soup = BeautifulSoup(HTMLFile, "html.parser")
html_text = soup.get_text()
f = open("html_text.txt", "w")         # Creating html_text.txt File
for line in html_text:
	f.write(line)
f.close()
```

```python
# "what are the most important things you learned" responses
c = []
table1 = bsText.find("table")#, {"title":"TheTitle"} )
for row in table1.findAll("tr"):
    c.append(row.text[1:-1])
important = pd.DataFrame(c, columns=['comment'])
important
```

```python
# "what contributed" responses
c = []
table2 = bsText.findAll('table')[1]
for row in table2.findAll("tr"):
    c.append(row.text[1:-1])
contri = pd.DataFrame(c, columns=['comment'])
contri
```

```python
# "please respond to the following" responses
row_names = []
mean = []
median = []
strongly_disagree = []
disagree = []
neutral = []
agree = []
strongly_agree = []

table3 = bsText.findAll('table')[2]
table_body = table3.find('tbody')

for h in table_body.find_all('th'):
  row_names.append(h.text)

for tr in table_body.find_all('tr'):
  td = tr.find_all('td')
  mean.append(td[0].text)
  median.append(td[1].text)
  strongly_disagree.append(td[2].text)
  disagree.append(td[3].text)
  neutral.append(td[4].text)
  agree.append(td[5].text)
  strongly_agree.append(td[6].text)

dict_eval = {'name': row_names, 'mean': mean, 'median': median, 'strongly_disagree': strongly_disagree, 'disagree': disagree, 'neutral': neutral, 'agree': agree, 'strongly_agree': strongly_agree}
eval = pd.DataFrame(dict_eval)
eval
```

```python
# "additional comments" responses
c = []
table3 = bsText.findAll('table')[3]
for row in table3.findAll("tr"):
    c.append(row.text[1:-1])
additional = pd.DataFrame(c, columns=['comment'])
additional
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# CONTENT
stopwords = nltk.corpus.stopwords.words('english')
# remove stopwords
important['stopwords_removed'] = important['comment'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
# remove punctuations
important["clean_comment"] = important['stopwords_removed'].apply(lambda x: re.sub('[^A-Za-z0-9]', ' ', x))

pd.DataFrame(pd.Series(' '.join(important['clean_comment']).lower().split()).value_counts()[:10])
```

```python
# Sentiment Analysis on "what contributed most" and "additional comments"
all_comments = pd.concat([contri, additional], axis=0)

# remove stopwords
stopwords = nltk.corpus.stopwords.words('english')
all_comments['stopwords_removed'] = all_comments['comment'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))

analyzer = SentimentIntensityAnalyzer()
scores = pd.DataFrame(all_comments['stopwords_removed'].map(lambda x: analyzer.polarity_scores(x)))

total_neg = 0
total_neu = 0
total_pos = 0
for i in range(0, len(scores) - 1):
  list_score = re.findall(r"\d+\.\d+", str(scores.iloc[i]))
  total_neg += float(list_score[0])
  total_neu += float(list_score[1])
  total_pos += float(list_score[2])
avg_neg = total_neg / len(scores)
avg_neu = total_neu / len(scores)
avg_pos = total_pos / len(scores)
[avg_neg, avg_neu, avg_pos]
```
